```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "temperature": [25, 30, 35, 28, 32, 33, 31],
    "sunlight_hours": [6, 8, 9, 7, 8, 9, 7],
    "panel_age": [2, 3, 4, 2, 5, 3, 4],
    "efficiency": [75, 80, 78, 76, 79, 81, 77]
}

df = pd.DataFrame(data)

plt.scatter(df["temperature"], df["efficiency"])
plt.title("Temperature vs Efficiency")
plt.show()

plt.bar(df["panel_age"], df["efficiency"])
plt.title("Panel Age vs Efficiency")
plt.show()

plt.hist(df["efficiency"])
plt.title("Efficiency Distribution")
plt.show()
```
